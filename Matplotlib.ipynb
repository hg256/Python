```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,11)
y = x **2
```

```python
x
```

```python
y
```

```python

plt.plot(x,y,'b--')
plt.xlabel('X_Label')
plt.ylabel('Y_Label')
plt.title('Title')
#plt.show()
```

```python
plt.subplot(221)
plt.plot(x,y,'b')
plt.subplot(222)
plt.plot(x,y**3,'y*')
plt.subplot(223)
plt.plot(x,y**2,'g')
plt.tight_layout()
```

## Object Oriented method


```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.45,0.5, 0.5])
axes.plot(x,y,'b', label = "Test")
axes2.plot(x,y**2,'g', label = "Green")

axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Linear')
axes.legend(loc=4)
axes2.legend(loc=0)
```

## Subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
plt.tight_layout()
#axes.plot(x,y)
for row_axs in axes:
    for current_ax in row_axs:
        current_ax.plot(x,y)
```

```python
fig, axes1 = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
axes1[0].plot(x,y,'g',label = 'Green')
axes1[1].plot(y,x,'b', label = 'Blue')
axes1[1].set_title('Blue')
axes1[0].set_title('Green')
axes1[0].legend()
axes1[1].legend()
```

## Figure size and DPI

```python
fig1 = plt.figure(figsize=(1,1),dpi=600,facecolor='g')
axes = fig1.add_axes([0,0,2.5,2.5])
axes.plot(x,y,'b', label = "Test")
#axes.set_xlim([0,4]) #setting limit on x and y axes
```

```python
fig1,axes = plt.subplots(nrows=2,ncols=1,figsize=(5,5))
axes[0].plot(x,y,color='purple',linewidth=3,alpha=0.5,linestyle='steps')
axes[1].plot(x,y,color='orange',lw=3,alpha=0.5,ls='-',marker='o',markersize=10,markerfacecolor='Yellow',markeredgewidth =2,markeredgecolor='green')
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.grid(True)

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

```python
n = np.array([0,1,2,3,4,5])
```

```python
fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(x, y + 0.25*np.random.randn(len(x)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
plt.tight_layout()
```
